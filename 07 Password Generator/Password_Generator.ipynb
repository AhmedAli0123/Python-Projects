```python
import random

print("Welcome to Password")

chars = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ@.?0123456789"  #Character of Password

amount_password=input("Wnter the amount of password: ")                      #Amount Of Password
amount_password=int(amount_password)

length_password=input("Wnter the length of password: ")                    #Length Of Password
length_password=int(length_password)


for x in range(amount_password):
    password="".join(random.sample(chars,length_password))
    print(password)
```
